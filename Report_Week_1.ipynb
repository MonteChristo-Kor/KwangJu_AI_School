##1. 자율주행 : **인공지능의 윤리학**

###1-1. 자율 주행 정의
  - 자율 주행은 인간이 운전에 대한 전체 개입 없이, 일부 개입, 감독 혹은 전혀 개입하지 않고 이동 장치를 AI에 의지해 작동하는 것
  - 보통 국제자동차기술자협회(SAE) 미국 도로교통안전국에서 이야기 하는 단계별 정의를 적용
    + 0단계 - 비자동화 : 자율 주행 기능이 **전혀 없는** 일반 이동 장치
    + 1단계 - 운전자 지원 기능 : 자동 브레이크, 자동 속도 조절 등 **운전을 보조**하는 수준
    + 2단계 - 부분적 자율 주행 : 운전자가 운전하는 상태에서 2가지 이상의 자동화 기능이 동시에 작동. 부분 자율주행으로 **운전자 상시 감독 필요**
    + 3단계 - 조건부 자율 주행 : 자동차 내 AI에 의한 제한적인 자율주행이 가능하나 특정 상황에 따라 **운전자의 개입이 반드시 필요**
    + 4단계 - 고급 자율 주행 : **시내 주행을 포함한 도로환경**에서 주행시 운전자 개입이나 모니터링이 필요하지 않은 상태
    + 5단계 - 완전 자동화 : **모든 환경** 하에서 운전자의 개입이 필요하지 않음
  - 현재 양산차의 자율주행 기술 완성 수준 : 2단계 내지는 3단계 수준

###1-2. 자율 주행이 이뤄져야 하는 이유
  - 사고 감소 
    + 한국의 2019년 교통사고 사상자는 345,061명
    + 2017년 통계 기준 사상자의 10.0%는 음주운전으로, 2.2%는 졸음운전으로 인해 발생
    + 자율 주행이 가능하다면 이러한 문제들로 인한 사상자를 줄일 수 있고, 그로 인해 발생되는 사회적 비용 역시 절감 가능
  - 이동 수단의 가치 상승
    + 기존의 이동 수단, 특히 자동차는 운전을 해야하므로 차 안에서 운전에 신경을 써야 하고, 동승자 역시 운전자의 운전을 방해하지 않도록 노력해야 함
    + 자율 주행 기술이 완성된다면 자동차는 업무적 공간, 혹은 문화적 공간, 휴식의 공간으로 재창조 될 수 있음
    + 차량이나 이동 수단의 소유가 아닌 목적에 따른 이동 수단을 결정함으로 이동 수단에 대한 새로운 패러다임 설정 가능
    + 특히 최근 코로나와 같이 전염성 높은 질병들이 유행할 때 다중 인원이 밀폐된 공간에서 영화를 보거나 공연을 보는게 아닌 야외 공간에서 이동수단이 하나의 관람 공간으로 변신 가능
  - 이동에 관한 Data가 수집되면서 다른 이동 수단에 대한 정교한 알고리즘 구현 가능
    + 인간이 운전을 하면 오감을 통해 받아들이는 정보나 개인적인 판단 준거는 개인의 머릿 속에서만 남지만 자율 주행 차로 변경되며 ** 그로 인해 얻어디는 다양한 Data를 선택적으로 활용** 가능

##2. 언어 : **언어 보존 및 고전 번역 알고리즘 개발**
###2-1. 인공지능과 언어 보존의 상관 관계
  - **언어는 두 번의 죽음을 맞는다 : 마지막 화자의 죽음, 그리고 기록물의 파괴**
    + 유네스코에 따르면 전 세계에 존재하는 언어는 6,000개
    + 세계 언어의 3분의 1은 사용하는 사람이 1,000명 미만이고, 43%는 소멸위기에 있으며, 매 2주마다 한 개꼴로 언어가 사라지고 있음. 
    + 1950년 이후부터 지금까지 약 230개의 언어가 소멸
    + 현재 소멸 위기에 처한 언어의 50% ~ 90%도 다음 세기면 사라질 위기
    + 언어는 문화와 정신의 연결 통로로 한 세대에서 다음 세대로 인간의 문명을 연결하는 이음줄
    + 언어를 잃는다는 것은 단순히 언어를 사용하지 못한다는 것이 아니라 다양한 문화에서 이룩한 정신들을 단절시키고 소멸시켜버림을 뜻함.

###2-2. 인공지능을 통해 언어 보존, 고전 번역을 해야 하는 이유 : **언어는 패턴이다**
  - **언어는 무에서 유로 갑자기 생겨나지 않는다.**
    + 언어는 그 집단에서 이미 사용하고 있는 말, 혹은 주변에서 사용하고 있는 언어나 말을 기반으로 생겨나므로 하나의 언어를 알고 있다는 것은 주변의 몇 가지 언어의 패턴을 유추할 수 있다는 의미
    + 언어는 주어, 동사, 목적어 등 모든 언어에서 일반적으로 보여지는 요소들이 있고, 그 요소들 간에 상관관계인 패턴이 있음. 
    + 즉, 수식이나 알고리즘으로 언어의 개괄적 패턴을 구성할 수 있고, 이는 이미 사멸되어 버린 언어들도 현재 언어들을 거슬러 올라가면 패턴을 분석해서 다시 살릴 수 있다는 의미
    + **언어를 다시 기억해낸다는 것은 언어가 가지고 있는 존재적 가치, 그리고 그 언어를 통해 유추할 수 있는 역사, 그 언어로 작성된 문화유산까지 기억해 낼 수 있는 매우 중요한 사건임**
  - 고전 번역도 패턴이다
    + 빠른 속도를 내는 것은 인간이 엔진이나 모터를 따라잡을 수 없고, 계산은 컴퓨터를 이길 수 없음
    + 언어간의 상관관계를 충분히 분석할 수 있는 Data가 있다면 알고리즘을 통한 번역이 인간의 번역보다 정확도와 효율성을 확보할 수 있음. 
    + 고전은 시대가 지나 사용하는 사람이 거의 없다는 점에서 사멸성 언어와 공통점을 보임
    + 전문가가 일생을 바쳐 1권 혹은 몇 권을 번역을 할 정도로 번역의 속도가 더디고 읽는 이가 많지 않아 많은 고전 언어 서적이 번역되지 않고 있음
    + 후대를 위해서 꾸준히 그 시대에 사용하는 언어로 번역을 해놓음으로 문화의 다양성 확보 가능

###2-3. 이 분야를 선택한 이유 : 역사는 컴퓨터의 저장과는 다른 총체적 정신 집합체의 유산이다.
  - 역사는 단순한 기록이 아니라 인간 전체의 사상, 정신, 문화를 담은 혼
    + 인공지능도 빅데이터를 통해 학습을 하고, 그로 인해 정확도를 올리듯이 인간도 다양한 역사에 대한 데이터와 지식으로 과거를 반성하고 미래를 올바르게 계획할 수 있음 
    + 현재에 가지고 있는 데이터는 인공지능 자체를 통해 삶을 편리하게 하지만 고전에서 잊혀진 지식들을 가져옴으로 인간 본연의 모습을 찾아 갈 수 있음
    * 인공지능이 복잡하고 어려운 일들을 자동화처리를 함으로 인간은 인간의 본연의 모습을 탐구해 나가는 분야로 더욱 매진해야하고 그러기 위해서는 저변 Data를 인공지능을 통해 확보 필요
  - 인공지능은 인간을 돕는 도구이다
    + 하기 싫은 일, 하기 어려운 일, 할 수 없는 일이 아니라 인간을 위한 일에 인공지능이 사용되어야 함
    + 그런 의미에서 인간의 역사에서 사라져가는 언어를 보존하고, 이미 사라져 버린 언어들을 기록으로 남기고, 남겨진 기록을 다시 현대어로 번역하는 일은 인간의 보조 도구로서 할 수 있는 최상의 보조 업무

    + 이는 개인들마다 선호하는 장면들이 다르기 때문에 맞춤형으로 제공이 되어야 했으나 제작하는데 들어가는 비용과 시간이 필요하므로 천편일률적인 하이라이트가 제공되어져 왔음
    + 인공지능을 통해 화면에 나오는 패턴을 분석하고 그것을 태그화하고 정리해서 소비자가 원하는 하이라이트 영상(이미지들의 집합)으로 구성한다면 영상에 대한 맞춤형 패러다임 창조 가능
  - 소비자 수요 중심으로 만들어진 하이라이트를 통해 전체 컨텐츠 소비 촉진 가능
    + 천편일률적으로 만들어진 하이라이트는 흥미를 유발하지 못하고 그로 인해 버려지는 Input이 늘어나게 됨
    + 인공지능을 통해 개인별 맞춤형 하이라이트를 생성할 수 있다면 버려지는 Input을 줄일 수 있고, 소비자 만족도 확보 가능

###4-2. 이 분야를 선택한 이유 : **낭비를 줄이면 효율적인 투자가 가능하다**
  - 시장을 빠르게 확보할 수 있고, 단지 낭비를 줄일 수 있는 것만으로 시장에 바로 진입 가능
    + 축구, 야구, 이종격투기 등 다양한 스포츠, 게임, 영화 등 시청하는 전 분야에 걸쳐 적용 가능
    + 이미 예고편, 미리보기, 소개 등 요약되어 컨텐츠가 만들어지고 있고, 그 컨텐츠를 통해 전체 컨텐츠를 소비할 것인지를 판단
    + 문제는 예고편 등이 제작될 때 비용이 높기 때문에 버려지는 Input이 많아져 쓸데 없는 비용 상승
    + 꼭 필요한 요소에 자본을 투자함으로 전체 Cost 및 수요 단가를 낮출 수 있고, 잉여 이익을 높일 수 있으므로 하이라이트 부분에 투자가 된다면 경제적 효과가 클 것으로 판단
  - 영화 등 기타 양질의 컨텐츠에서 보고 싶어하는 장르 뿐만 아니라 장면들을 모아서 하나의 컨텐츠를 만들 수 있는 새로운 장르 개척 가능
    + 철저히 소비자 중심으로 컨텐츠를 만들 수 있는 기술로 개발 가능
